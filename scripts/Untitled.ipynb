```python
from clickhouse_driver import Client

# Conectamos al servidor ClickHouse (nombre del servicio en docker-compose)
# Si 'clickhouse' no funciona, prueba 'clickhouse_server' o 'localhost'
client = Client(host='clickhouse') 

print("--- Verificando Datos en ClickHouse ---")

# 1. Contar registros
conteo = client.execute("SELECT count(*) FROM dw_analitico.ventas_resumen")[0][0]
print(f"Total de filas resumidas: {conteo}")

# 2. Ver las primeras 5 filas
print("\nEjemplo de datos (Top 5):")
filas = client.execute("SELECT * FROM dw_analitico.ventas_resumen LIMIT 5")
for fila in filas:
    print(fila)
```
